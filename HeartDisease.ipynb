```python
#finding the missing values(null values)
null = df.isnull().sum() #storing value in null variable
print("Null values in dataset are ",null)
```

```python
#dropping the null values : columns(entire bmi column will be deleted: axis=1)
df.dropna(axis=1) 
```

```python
#rows with null values will be deleted
df.dropna()
```

```python
#to fill null values
#default method of filling using 0
df.fillna(0)
```

```python
#filling null values with forward and backward fill

#forward fill: fills the NaN with previous value
df.fillna(method = "ffill")
```

```python
#backward fill: fills the NaN value with next value
df.fillna(method = 'bfill')
```

```python
#filling NaN values with mean
df.fillna(df['bmi'].mean())
```

```python
#filling NaN values with median
df.fillna(df['bmi'].median())
```

```python
#filling NaN values with mode(mode doesn't replace NaN vales)
df.fillna(df['bmi'].mode())
```

```python
#dropping unnecessary columns

#drop method
df.drop('Residence_type', axis=1)
df.drop(columns=['age','gender'], axis=1)
```

```python
import numpy as np
a = np.array(df.age)

plt.xticks([0,25,50,75,100])
plt.xlabel("age",
          fontdict={
                "family":"Comic Sans MS",
                "color":"black",
                "weight":"bold",
                "size":15
})
plt.ylabel("id", 
           fontdict={
                "family":"Comic Sans MS",
                "color":"black",
                "weight":"bold",
                "size":15
})
plt.title("HISTOGRAM", 
          fontdict={
                "family":"Comic Sans MS",
                "color":"black",
                "weight":"bold",
                "size":20
})

plt.hist(a, bins=[0,25,50,75,100], color='pink')
```

```python
fig = plt.figure()

work = np.array(df.work_type)
ids = np.array(df.age)

plt.xlabel("WorkType", 
           fontdict={
                "family":"Comic Sans MS",
                "color":"black",
                "weight":"bold",
                "size":15
})
plt.ylabel("ID", 
           fontdict={
                "family":"Comic Sans MS",
                "color":"black",
                "weight":"bold",
                "size":15
})
plt.title("WORK", 
          fontdict={
                "family":"Comic Sans MS",
                "color":"black",
                "weight":"bold",
                "size":20
})

plt.bar(work,ids, color='pink')
plt.show()
```
